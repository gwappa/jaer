```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tab = pd.read_csv("MeanTracker_12092017163613.log")
tab
```

```python
ts1 = np.array(tab.firstts)
dts = np.diff(ts1)
```

```python
plt.hist(dts/1000, bins=100, range=(0, 6), color='k', edgecolor='none')
plt.title("firstts intervals")
plt.xlabel("interval (ms)")
plt.ylabel("count")
plt.show()
```

```python
plt.hist(dts/1000, bins=100, range=(0, 0.5), color='k', edgecolor='none')
plt.title("firstts intervals")
plt.xlabel("interval (ms)")
plt.ylabel("count")
plt.show()
```
